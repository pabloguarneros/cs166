```python
import numpy as np
import matplotlib.pyplot as plt
x = y = np.array([-2,-1,0,1,2])
### meshgrid
# interprets the x inputs as rows
# interprets the y inputs as columns
x_grid, y_grid = np.meshgrid(x, y)
```

```python
vector_x = x_grid + y_grid
vector_y = x_grid - y_grid
```

```python
plt.figure(figsize=(6, 6))
plt.quiver(x_grid, y_grid, vector_x, vector_y)
plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 2.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Create grid coordinates
x = np.linspace(-2 * np.pi, 2 * np.pi, 20)
y = np.linspace(-np.pi, np.pi, 10)
x_grid, y_grid = np.meshgrid(x, y)
# Compute vector field
vector_x = y_grid
vector_y = -np.sin(x_grid)
# Plot vector field
plt.figure(figsize=(8, 4))
plt.quiver(x_grid, y_grid, vector_x, vector_y)
plt.show()

# Make A Stream Plot
plt.figure(figsize=(8, 4))
plt.streamplot(x_grid, y_grid, vector_x, vector_y)
plt.show()
```

```python
import matplotlib.pyplot as plt
# Create grid coordinates
j = np.linspace(0,1, 20)
i = np.linspace(0,1, 20)
i_grid, j_grid = np.meshgrid(i, j)
# Compute vector field
alpha = 0.03
w = 0.9
vector_i = i_grid + alpha*w*(i_grid-j_grid)
vector_j = j_grid - alpha*w*(i_grid-j_grid)
# Plot vector field
plt.figure(figsize=(8, 4))
plt.quiver(i_grid, j_grid, vector_i, vector_j)
plt.show()
```
