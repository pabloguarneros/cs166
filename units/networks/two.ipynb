```python
from pylab import *
import networkx as nx


subplot(4, 1, 1)
g = nx.complete_graph(4)
nx.draw(g, with_labels = True)
title('Complete graph showing four nodes')

subplot(4, 1, 2)
g2 = nx.complete_graph(5)
nx.draw(g2, with_labels = True)
title('Complete graph showing five nodes')

subplot(4, 1, 3)
g3 = nx.complete_bipartite_graph(2, 3)
nx.draw(g3, with_labels = True)
title('complete (where all possible intergroup edges are present) bipartite graph')

subplot(4, 1, 4)
g4 = nx.complete_bipartite_graph(3, 3)
nx.draw(g4, with_labels = True)
title('complete (where all possible intergroup edges are present) bipartite graph')

show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

fig, ax = plt.subplots(4, 1, figsize=(6, 6))
g = nx.hypercube_graph(4)
nx.draw(g, with_labels = True, ax=ax)
title('a hypercube graph of four dimensions')

g2 = nx.ladder_graph(5)
nx.draw(g2, with_labels = True, ax=ax)
title('a ladder graph of length five')

g3 = nx.barbell_graph(20, 20)
nx.draw(g3, with_labels = True, ax=ax)
title('a barbell graph made of two 20-node complete graphs connected by a single edge')

g4 = nx.wheel_graph(100)
nx.draw(g4, with_labels = True, ax=ax)
title('a wheel graph made of 100 noes')

plt.show()
```
